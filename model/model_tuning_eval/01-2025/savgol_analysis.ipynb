```python
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
pre_savgol_df = pd.read_csv('2024_proj_pre_savgol_filtering.csv')
pre_savgol_df['GP'] = pre_savgol_df['GPprb'] * 82
pre_savgol_df['G/60'] = pre_savgol_df['Gper1kChunk']/1000*2 * 60
pre_savgol_df['A1/60'] = pre_savgol_df['A1per1kChunk']/1000*2 * 60
pre_savgol_df['A2/60'] = pre_savgol_df['A2per1kChunk']/1000*2 * 60
pre_savgol_df['P/60'] = pre_savgol_df['G/60'] + pre_savgol_df['A1/60'] + pre_savgol_df['A2/60']
pre_savgol_df['G'] = pre_savgol_df['G/60']/60 * pre_savgol_df['GP'] * pre_savgol_df['ATOI']
pre_savgol_df['A'] = (pre_savgol_df['A1/60'] + pre_savgol_df['A2/60'])/60 * pre_savgol_df['GP'] * pre_savgol_df['ATOI']
pre_savgol_df['P'] = pre_savgol_df['P/60']/60 * pre_savgol_df['GP'] * pre_savgol_df['ATOI']
pre_savgol_df = pre_savgol_df.sort_values(by='P', ascending=False)
pre_savgol_df = pre_savgol_df.drop(columns=['GPprb', 'Gper1kChunk', 'A1per1kChunk', 'A2per1kChunk'])
pre_savgol_df
```

```python
post_savgol_df = pd.read_csv('2024_proj_post_savgol_filtering.csv')
post_savgol_df['GP'] = post_savgol_df['GPprb'] * 82
post_savgol_df['G/60'] = post_savgol_df['Gper1kChunk']/1000*2 * 60
post_savgol_df['A1/60'] = post_savgol_df['A1per1kChunk']/1000*2 * 60
post_savgol_df['A2/60'] = post_savgol_df['A2per1kChunk']/1000*2 * 60
post_savgol_df['P/60'] = post_savgol_df['G/60'] + post_savgol_df['A1/60'] + post_savgol_df['A2/60']
post_savgol_df['G'] = post_savgol_df['G/60']/60 * post_savgol_df['GP'] * post_savgol_df['ATOI']
post_savgol_df['A'] = (post_savgol_df['A1/60'] + post_savgol_df['A2/60'])/60 * post_savgol_df['GP'] * post_savgol_df['ATOI']
post_savgol_df['P'] = post_savgol_df['P/60']/60 * post_savgol_df['GP'] * post_savgol_df['ATOI']
post_savgol_df = post_savgol_df.sort_values(by='P', ascending=False)
post_savgol_df = post_savgol_df.drop(columns=['GPprb', 'Gper1kChunk', 'A1per1kChunk', 'A2per1kChunk'])
post_savgol_df
```

```python
target_df = pd.read_csv('2023-2024_skater_data.csv')
target_df = target_df[['PlayerID', 'Player', 'Team', 'Position', 'GP', 'TOI', 'Goals', 'First Assists', 'Second Assists', 'Total Points']]
target_df['ATOI'] = target_df['TOI']/target_df['GP']
target_df['G/60'] = target_df['Goals']/target_df['ATOI']/target_df['GP']*60
target_df['A1/60'] = target_df['First Assists']/target_df['ATOI']/target_df['GP']*60
target_df['A2/60'] = target_df['Second Assists']/target_df['ATOI']/target_df['GP']*60
target_df['P/60'] = target_df['Total Points']/target_df['ATOI']/target_df['GP']*60
target_df['G'] = target_df['Goals']
target_df['A'] = target_df['First Assists'] + target_df['Second Assists']
target_df['P'] = target_df['Total Points']
target_df = target_df.sort_values(by='P', ascending=False)
target_df = target_df.drop(columns=['TOI', 'Goals', 'First Assists', 'Second Assists'])
target_df
# target_df
```

```python
results = target_df.copy()
results = results.merge(pre_savgol_df[['PlayerID', 'GP', 'ATOI', 'G/60', 'A1/60', 'A2/60', 'P/60', 'G', 'A', 'P']], on='PlayerID', how='inner', suffixes=('', '_presavgol'))
results = results.merge(post_savgol_df[['PlayerID', 'GP', 'ATOI', 'G/60', 'A1/60', 'A2/60', 'P/60', 'G', 'A', 'P']], on='PlayerID', how='inner', suffixes=('', '_postsavgol'))
results = results[results['GP'] > 50]
results
```

```python
# Define the stat columns to compare
stat_columns = [
    "P/60", "ATOI", "G/60", "A1/60", "A2/60", "G", "A", "P", "GP"
]

# Create a dictionary to store MSE results
mse_results = {"Stat": [], "MSE Projections 1": [], "MSE Projections 2": []}

# Compute MSE for each stat
for stat in stat_columns:
    mse1 = mean_squared_error(results[stat], results[f'{stat}_presavgol'])
    mse2 = mean_squared_error(results[stat], results[f'{stat}_postsavgol'])
    
    mse_results["Stat"].append(stat)
    mse_results["MSE Projections 1"].append(mse1)
    mse_results["MSE Projections 2"].append(mse2)

mse_table = pd.DataFrame(mse_results)
mse_table
```

```python
# do the same but with MAE
stat_columns = [
    "P/60", "ATOI", "G/60", "A1/60", "A2/60", "G", "A", "P", "GP"
]

# Create a dictionary to store MAE
mae_results = {"Stat": [], "MAE Projections 1": [], "MAE Projections 2": []}

# Compute MAE for each stat
for stat in stat_columns:
    mae1 = mean_absolute_error(results[stat], results[f'{stat}_presavgol'])
    mae2 = mean_absolute_error(results[stat], results[f'{stat}_postsavgol'])
    
    mae_results["Stat"].append(stat)
    mae_results["MAE Projections 1"].append(mae1)
    mae_results["MAE Projections 2"].append(mae2)

mae_table = pd.DataFrame(mae_results)
mae_table
```
